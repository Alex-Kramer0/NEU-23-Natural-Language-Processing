(python) Create a demonstration in a jupyter notebook of changing a temperature hyperparameter. To do this, follow the steps below:

You should end up with a well-formed probability distribution. This probability distribution should reflect reasonable probabilities for your vocabulary given your input sequence.
$$ \texttt{softmax}(z_i) = \frac{e^{z_i}}{\sum_{j = 1}^k e^{z_j}}$$

#### Helpful Note 
Definition of Temperature: "Temperature is a number used to tune the degree of randomness. A lower temperature means less randomness; a temperature of 0 will always yield the same output. " - https://docs.cohere.com/docs/temperature

```python
# importing the necessary packages
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def soft_max(z, temp=1):
    """
    param z: a list of z values which are a float number between -1 and 1 representing probabilities
    param temp: takes in an int; represents temperature; default = 0; lower number = less random 
    func: takes in a list of z values and normalizes them using the formula provided
    return: a list of outputs (the z values normalized)
    """
    denominator = 0
    e = math.e
    
    # calculate denominator 
    for j in range(len(z)): # from 0 - length of z list 
        # for each item in the list of z score
        x = e**(z[j] / temp)
        # raise e to its correlated z score divided by temp
        denominator += x
        # add x to the denmoniator 
    
    out = [] # create empty list to append output values too 
    
    for i in range(len(z)):
        # loop through each z score in the list 
        numerator = e**(z[i] / temp)
        # create our numerator 
        out.append(numerator / denominator)
        # append our numerator / denominator to our out list 
    
    return out
```

```python
def sample(vocab, z, temp=1, top_k=-1):
    """
    param vocab: the list of vocabulary words
    param z: a list of z values which are a float number between -1 and 1 representing probabilities
    param temp: takes in an int; represents temperature; default = 0; lower number = less random 
    param top_k: the number of top words to consider during sampling
    func: takes in a list of z values and normalizes them using the formula provided
    return: returns a single word from the vocab list 
    """

    probabilities = soft_max(z, temp)
    # calls the soft_max function with z and temp; soft_max
    # calculates the probablities of z and returns a list of numbers

    if top_k != -1: # checks regular sampling or irregular
        output = np.random.choice(vocab, p=probabilities)
        # uses random to choose a random word & it's probability

    else:  # regular sampling 
        vocab_prob_tuples = [(word, prob) for word, prob in zip(vocab, probabilities)] 
        ## used LLM to help format how to create a tuple
        
        sorted_tuples = sorted(vocab_prob_tuples, key=lambda x: x[1], reverse=True)
        # creates a list of tuples, grouping each word to their probability & reverse sorts the list
        
        top_k_items = sorted_tuples[:top_k]
        # create a list of the top_k'th items in the list using slicing 
        
        
        probabilities_top_k = [0] * len(vocab) # create a list of 0's the size of the vocab list
        for word, prob in top_k_items:
            # for each word, prob in the top_k_items
            index = vocab.index(word)
            probabilities_top_k[index] = prob
            # find the index & set the index in the new list to the probability
            # zeroing out the non-top-k words
        
        output = np.random.choice(vocab, p=probabilities_top_k)
        # uses random to choose a random word from our top-k words
    
    return output
```

```python
# create input sequence 
inp = ['The cat slept']

# define a vocabulary, k len of vocab 
vocab = ['I', 'hello', 'world', 'the', 'cat', 'slept', 'in', 'sun', 
         'run', 'to', 'store', 'get', 'groceries', 'make', 'dinner']

# define z 
z = [0.4, 0.2, 0.1, 0.9, 0.2, 0.2, 0.5, 0.3, 0.3, 0.8, 0.5, 0.6, 0.4, 0.5, 0.4]
z_prob = soft_max(z, 1)

# change these to tune the parameters
top_k = 4
temp = 1


result = sample(vocab, z_prob, temp, top_k)

# printing results
print('The vocabulary words are: ', vocab, '\n\n')

print('The temperature is: ', temp, '\n')
print('The top_k is: ', top_k, '\n\n')

print('The Top-', top_k, ' distribution are: ', sample(vocab, z_prob, temp, top_k), '\n')
print('The next word selected is: ', result)
```

```python
# call soft_max 
temp_below_1 = soft_max(z, -2)
temp_1 = soft_max(z, 1)
temp_above_1 = soft_max(z, 2)

# plot the different outputs
plt.figure(figsize=(10, 6))
plt.plot(z, temp_below_1, label='Temperature -2')
plt.plot(z, temp_1, label='Temperature 1')
plt.plot(z, temp_above_1, label='Temperature 5')

# formatting
plt.title('Soft-max Output for Varying Temperatures')
plt.xlabel('Assigned z values')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```
